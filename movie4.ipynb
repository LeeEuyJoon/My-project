# '2019 2분기'의 데이터는 0 임을 적용시켜야 한다 이제

```python
# 날짜 범주화는 성공했으니 이제 이 데이터로 분기별 영화 시청 횟수 그랳프 다시 그려보자

date_part = df.groupby('날짜 분기별').count()
date_part['제목']

# 생각해보니까 저번에 그런 문제가 있었다 2021년 2분기에 영화를 안봐서 '날짜 분기별' 데이터에
# 아무것도 없었던..... 해결 할 수 있을까 리스트를 직접 작성하는거 말고
```

```python
# 딕셔너리를 이용하면 해결 할 수 있을지도...

year = [2019,2020,2021]
quater1234 = [1,2,3,4]

date_part.index[0][:4] == '2019'

date_dict = {}

for aa in range(len(year)):
    for bb in range(len(quater1234)):
        date_dict[str(year[aa]) + " " + str(quater1234[bb]) + "분기"] = 0

date_dict

# 일단 value 값이 전부 0이고 key 값을 날짜 분기 순서대로 하는 딕셔너리 생성을 성공했다.
# 이제 date_part['제목']의 인덱스와 date_dict의 key값을 비교한 다음 일치하면 
# date_part['제목']의 데이터를 넣어주고 아니면 건너가고 이런식으로 하면 될 것 같다.
# 근데 크기가 다른데 되는거 맞나?
```

```python
keys_list = list(date_dict.keys())

# if keys_list[0] == date_part['제목'].index[0]:
#     date_dict[keys_list[0]] = date_part['제목'][0]

for aa in range(len(date_dict)):
    for bb in range(len(date_part.index)):
        if keys_list[aa] == date_part.index[bb]:
            date_dict[keys_list[aa]] = date_part['제목'][bb]

date_dict

# 와 드디어 됬다 이제 이걸로 그래프 그리기만 하면 끝난다...하...
```

```python
plt.plot(date_dict.keys(),date_dict.values())
# 아 다행히 그래프 잘 나온다 이제 서식만 적용하면 되겠다.
```

```python
# 서식만 적용하면 끝

plt.figure(figsize = (11,7))
plt.plot(date_dict.keys(),date_dict.values(),marker='D',markersize = 15,
         linewidth = 5,markeredgecolor = 'w',markeredgewidth = 4,
         color = 'orange')
plt.xticks(rotation=45)
plt.title('분기별 영화 시청 횟수')
plt.grid()
plt.show()
```
