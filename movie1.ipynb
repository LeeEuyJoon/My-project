# 내가 본 영화들의 장르별 비율

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window 한글 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 글자가 꺠지는 현상 해결
```

```python
df = pd.read_excel('./movie.xlsx',index_col='번호')
df
```

```python
df.groupby('장르').size()
```

```python
genre_count = [4,5,20,24,2,1,17,5,9,21,2,13,2]
genre_name = ['SF','공포','드라마','로맨스','모험','미스터리','범죄','스릴러','애니메이션','액션','전쟁','코미디','판타지']
explode = [0.05]*13

plt.pie(genre_count,labels = genre_name,autopct = '%.1f%%',explode=explode)
plt.show()
```

```python
# 랜덤 색상 코드 추출하기

import numpy as np
color = list(np.random.choice(range(256),size=6))
print(color)

import random
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])]
print(color)
```

```python
colors = ['#09B068','#DB8490','#D28A15','#439D7F','#2CCFFD','#119DDF','#52B5D1','#ADF5B8','#B025A7','#D5E92E','#C52532','#4876FF','#E18D19']

plt.pie(genre_count,labels = genre_name,autopct = '%.1f%%',explode=explode,colors=colors)
plt.show()
```

```python
colors=['lightgray','lightsalmon','greenyellow','plum','gold','lightsteelblue','royalblue','palegreen','mediumorchid','lavender','rosybrown','navajowhite','powderblue']

plt.figure(figsize=(8,8))
plt.pie(genre_count,labels = genre_name,autopct = '%.1f%%',explode=explode,colors = colors)
plt.legend(loc=(1.2,0.1),prop={'size':17})
plt.show()
```

```python
wedgeprops = {'width':0.7,'edgecolor':'w','linewidth':5}

plt.rc('font',size=20)
plt.figure(figsize=(8,8))
plt.pie(genre_count,labels = genre_name,autopct = '%.1f%%',colors = colors,wedgeprops=wedgeprops)
plt.legend(loc=(1.2,0.1),prop={'size':17},title='장르')
plt.show()
```
