# 내가 본 영화중 같은 감독의 영화들 살펴보기

```python
import pandas as pd

df = pd.read_excel('./movie.xlsx',index_col='번호')
df
```

```python
director = df.groupby('감독').count()['제목']
director.sort_values(ascending = False)
```

```python
director_list = []

for x in range(director.size):
    if director[x] > 1:
        director_list.append(director.index[x])

director_list
```

```python
director_dict = {}

for x in range(len(director_list)):
    director_dict[director_list[x]] = 0
    
director_dict
```

```python
for x in range(len(df)):
    for y in range(len(director_list)):
        if df['감독'][x+1] == director_list[y]:
            if director_dict[director_list[y]] == 0:
                director_dict[director_list[y]] = [df['제목'][x+1]]
            else:
                director_dict[director_list[y]].append(df['제목'][x+1])            
director_dict
```

```python
director_list = list(director_dict.keys())

max = 0
for x in range(len(director_list)):
    if len(director_dict[director_list[x]]) > max:
        max = len(director_dict[director_list[x]])
max
```
